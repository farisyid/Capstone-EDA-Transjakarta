## **Overview** 

TransJakarta adalah sebuah sistem transportasi Bus Rapid Transit (BRT) pertama di Asia Tenggara dan Selatan dengan jalur lintasan terpanjang di dunia (208 km). Sistem BRT ini didesain berdasarkan sistem TransMilenio di Bogota, Kolombia. Terhitung sejak 1 Februari 2004, TransJakarta resmi beroperasi. 

Pada tahun 2011, TransJakarta menenerapkan sistem manajemen armada terpadu. Salah satunya, melakukan integrasi dengan operator bus untuk menyediakan layanan feeder Busway bagi para pelanggannya. Di tahun ini juga TransJakarta mulai melakukan persiapan pembukaan layanan koridor 11 & 12 serta mengganti bus koridor 1 dengan bus articulated.

TransJakarta mempersiapkan pembukaan koridor - koridor baru seperti, persiapan tender operator, pembangunan jalur Busway, ramp, halte dan infrastruktur pendukung lainnya. Di tahun 2013, TransJakarta juga mulai menerapkan sistem e-ticketing untuk seluruh koridor serta integrasi sistem e-ticketing dengan feeder Busway dan angkutan umum lainnya.

Pada 27 Maret 2014, TransJakarta berubah status menjadi bentuk BUMD (Badan Usaha Milik Daerah) dan resmi berganti nama menjadi PT. Transportasi Jakarta. Rencana pembukaan layanan koridor-koridor baru terus berjalan, inovasi-inovasi pun diluncurkan demi kenyamanan para pelanggan.

Pada April 2016, Transjakarta resmi mengoperasikan bus khusus wanita yang berwarna pink dalam rangka memperingati hari kartini. Bus khusus wanita ini dapat ditemui di koridor 1 (Blok M - Kota). Transjakarta juga akan menambah bus khusus wanita di koridor - koridor lainnya.

Site Visit Transjakarta: Senior Traveler pertama kali diadakan. Peserta untuk kegiatan pertama kali diikuti oleh 40 lansia dari Panti Werdha Sasana Bina Mulya. Kegiatan ini berjalan rutin setiap minggunya dan diikuti oleh peserta yang berumur 60-70 tahun dari berbagai panti sosial. Harapannya ke depan, Transjakarta tidak hanya nyaman dan aman bagi pelanggan pada umumnya tetapi juga menjadi angkutan umum yang ramah dan mudah bagi anak-anak kecil maupun pelanggan Lansia.

Ada 2 cara untuk menangani *missing value*:
* pertama, menghapus baris/kolom yang berisi *missing value*. Cara ini tidak disarankan dalam kasus kita, karena *missing value* masih bisa di isi dengan bagian dari kolom lainnya.
* kedua, mengisi data yang hilang. Cara ini lebih disarankan. Ada beberapa metode yang bisa digunakan untuk mengisi missing value, cara yang paling baik adalah dengan mengisi data yang hilang dengan nilai sebenarnya, atau sedekat mungkin dengan nilai asli dan jika data memilki type object, data masih bisa di isi dengan yang berkaitan dengan data dari kolom yang lain contohnya seperti `corridorID` dengan `corridorName` yang masih berkesinambungan, jadi jika di `corridorID` kosong, bisa mengambil value yang sama di baris yang lain, begitu juga `corridorName`.

#### **Fixing Missing Value**

mengisi kolom `corridorID` dengan .filna(method='ffill') yang berfungsi untuk mengisi kan *missing value* berdasarkan data yang berkesinambungan dengan kolom `corridorName` diatas data *missing value* itu sendiri. Begitu juga sebaliknya dengan kolom `corridorName`.

```python
# Kesinambungan antara kolom `corridorName` dan `corridorID` dengan `corridorID` sebagai pivot missing value nya
tj.loc[(tj['corridorName'] == 'Kampung Rambutan - Blok M') | (tj['corridorID'].isna()) & (tj['corridorID']=='7B'), ['corridorName', 'corridorID']]
```

Dari 2 tabel diatas, bisa disimpulkan semua 'corridorName' dan 'corridorID' memiliki keterkaitan, jadi missing value bisa di isi dengan .fillna(method='ffill') yang fungsinya mengisi missing value dari data pattern yang sudah ada dengan groupby corridorID sebagai acuan

```python
# Penggunaan groupby untuk pola data dari kolom `corridorID` dan `corridorName`, lalu menggunakan transform untuk pengaplikasian lambda .fillna metode ffill dan bfill kepada groupby tadi berdasarkan `corridorName`
tj['corridorName'] = tj.groupby('corridorID')['corridorName'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

## **Recomendation**

Dari analisa diatas, didapat lah insight-insight yang bisa menjawab permasalahan PT. TransJakarta dalam konsistensi `mengatasi kemacetan dan peningkatan fasilitas halte-halte`. <br>
Berdasarkan insight-insight diatas dapat dijadikan acuan dalam membuat rekomendasi yang bisa dilakukan sebagai berikut :
1. Dari rute-rute yang mendapati banyak penumpang bisa di buatkan lagi rute baru untuk pengalihan kepadatan rute tersebut yang lebih efisien.
2. Dari rute-rute yang mendapati banyak penumpang bisa dibuatkan alternatif jalan layang baru untuk mempercepat mobilisasi penumpang.
3. Untuk mengatasi kemacetan diJakarta perlu dianalisis secara global, berapa porsi transportasi umum milik PT. Transjakarta pada kemacetan Jakarta.
4. Dari halte-halte yang memiliki kepadatan penumpang bisa ditambahkan halte bayangan agar tidak ada penumpukan antrian.
5. Dari halte-halte yang memiliki penumpang yang banyak itu bisa dijadikan inovasi untuk bekerja sama dengan Transportasi Umum Jakarta lainnya dengan membangun gabungan dari halte dan stasiun LRT atau MRT agar semakin terintegerasi satu sama lain.
6. Dari halte-halte yang padat penumpang bisa dibuatkan jasa advertisement untuk penambahan laba PT. Transjakarta.
7. Dari halte-halte yang padat penumpang yang didominasi oleh wanita, agar ditambah lagi jalur khusus wanita, begitupun sedikit porsi untuk lansia agar mereka tetap nyaman dalam penggunaan transportasi umum PT. TransJakarta.
8. Apabila rekomendasi diatas dijalankan dengan baik, itu bisa meningkatkan pengguna transportasi umum PT. Transjakarta dan meningkatkan income perusahaan.
